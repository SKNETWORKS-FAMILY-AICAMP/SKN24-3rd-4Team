```python
from langchain_chroma import Chroma
from shared_embedding import get_embedding_model

ROOT = r'C:\skn24\team_project_3\Insurance_Benefit_chatbot\Insurance_Benefit_Chatbot'

embedding = get_embedding_model()

checks = [
    ("Bupa latest",      ROOT + r'\vectordb\bupa\bupa_latest',          'bupa_latest'),
    ("Bupa all",         ROOT + r'\vectordb\bupa\bupa_all',              'bupa_all'),
    ("Cigna all",        ROOT + r'\vectordb\cigna',                      'cigna_all'),
    ("Allianz latest",   ROOT + r'\vectordb\allianz\allianz_latest',     'allianz_latest'),
    ("Allianz all",      ROOT + r'\vectordb\allianz\allianz_all',        'allianz_all'),
    ("Tricare text",     ROOT + r'\vectordb\tricare\tricare_text',       'tricare_rag'),
    ("Tricare table",    ROOT + r'\vectordb\tricare\tricare_table',      'tricare_cost_tables'),
]

for label, path, collection in checks:
    try:
        db = Chroma(collection_name=collection, embedding_function=embedding, persist_directory=path)
        count = db._collection.count()
        print(f"{'✅' if count > 0 else '❌ 비어있음'} [{label}] '{collection}' → {count}개")
    except Exception as e:
        print(f"❌ [{label}] 오류: {e}")
```

```python
import os

ROOT = r'C:\skn24\team_project_3\Insurance_Benefit_chatbot\Insurance_Benefit_Chatbot'

for insurer in ['cigna', 'tricare']:
    path = os.path.join(ROOT, 'vectordb', insurer)
    print(f"\n📂 {insurer}:")
    if os.path.exists(path):
        for root, dirs, files in os.walk(path):
            level = root.replace(path, '').count(os.sep)
            indent = '  ' * level
            print(f"{indent}{os.path.basename(root)}/")
    else:
        print(f"  ❌ 폴더 자체가 없음: {path}")
```
